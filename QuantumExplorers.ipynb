```python
qreg7=QuantumRegister(2,"Q_Reg")
creg7=ClassicalRegister(2,"C_Reg")
qc7=QuantumCircuit(qreg7,creg7)

qc7.h(qreg7[0])
qc7.cx(qreg7[0],qreg7[1])

qc7.draw(output='mpl')
```

```python
qc7.measure(qreg7,creg7)
job=execute(qc7,simulator)
counts=job.result().get_counts()
print(counts)
```

### Circuit B is correct
